```python
import pandas as pd
```

```python

import matplotlib.pyplot as plt

# Read the CSV file with column names
df = pd.read_csv('pnl_10000_500.csv', names=['strat_name', 'time', 'pnl'])

# Modify the x label
df['time'] = df['time'].str.split('.').str[0]

# Plot the last column as pnl
ax = df.plot(x='time', y='pnl', title='PnL 2019-Jan-30', rot=45)
ax.tick_params(axis='x', labelsize=6)

plt.show()

```

```python
# Read the CSV file with column names
df = pd.read_csv('pnl_0502_0506.csv')

# Modify the x label
df['Time'] = df['Time'].str.split('.').str[0]

# Plot the last column as pnl
ax = df.plot(x='Time', y='Cumulative PnL', title='PnL 2022-May-02 to 2022-May-06', rot=45)
ax.tick_params(axis='x', labelsize=6)

plt.show()
```
